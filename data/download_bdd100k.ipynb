```python
import kagglehub
from pathlib import Path

# Download to current project directory instead of default cache
project_dir = Path.cwd()
out_dir = project_dir / "data" / "kaggle_datasets"
out_dir.mkdir(parents=True, exist_ok=True)

path = kagglehub.dataset_download(
    "solesensei/solesensei_bdd100k",
    force_download=False,
    output_dir=str(out_dir),
)

print("Path to dataset files:", path)
```

```python
from pathlib import Path
from huggingface_hub import snapshot_download

out_dir = Path.cwd() / "data" / "acdc_hf"
out_dir.mkdir(parents=True, exist_ok=True)

print(out_dir)


data_dir = snapshot_download(
    repo_id="mathpluscode/ACDC",
    repo_type="dataset",
    allow_patterns=["*.nii.gz", "*.csv"],
    local_dir=str(out_dir),              # download into current project
    local_dir_use_symlinks=False,        # actual files, not symlinks
)

print("Downloaded to:", data_dir)
```
